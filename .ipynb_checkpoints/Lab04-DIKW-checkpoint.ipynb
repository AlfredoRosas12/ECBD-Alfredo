```python
#Cargamos el Dataset
df = pd.read_csv("test.csv")
```

```python
#Limpieza
#Nombres de columnas en minusculas
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
#Detectar valores nulos
df.isnull().sum()
```

```python
#Eliminamos la columna "cabin"
df.drop(columns=["cabin"], inplace=True)
```

```python
df.columns
```

```python

df["age"] = df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
#Rellenamos la columna Fare con el valor más repetido (moda)
df["fare"] = df["fare"].fillna(df["fare"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#Validar datos 
df["age"].min()
```

```python
df["age"].max()
```

```python
df
```

```python
#Graficamos el total de pasajeros por clase
plt.figure(figsize=(8,4)) #Tamaño
sns.countplot(df, x="pclass")
plt.title("Total de pasajeros por clase")
plt.xlabel("Clase")
plt.ylabel("Total")
plt.show()
```

```python
#Graficamos las clases por genéro
plt.figure(figsize=(8,4))
sns.countplot(df, x="sex", hue="pclass")
plt.title("Total de Genéro por Clases")
plt.xlabel("Genéro")
plt.ylabel("Total")
plt.show()
```

```python
#Mostramos la edad por genéro y clase
plt.figure(figsize=(6,4))
sns.violinplot(df,x="sex", y="age", hue="pclass")
plt.title("Rando de edad por genéro y clase")
plt.xlabel("Genéro")
plt.ylabel("Edad")
plt.show()
```

```python
#CONOCIMIENTO
#Los pasajeros de primera clase pagaban tarifas más altas.
df.groupby("pclass")["fare"].mean()
```

```python
#La distribución muestra que había más hombres que mujeres
df["sex"].value_counts()
```
