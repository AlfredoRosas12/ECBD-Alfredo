```python
#----3. Análisis descriptivo
#Generar al menos 4 gráficas
```

```python
#Graficamos la cantidad de titulos por tipo
plt.figure(figsize=(8,4))
sns.countplot(df, x="TYPE")
plt.title("Cantidad de títulos por tipo")
plt.xlabel("Tipo de Contenido")
plt.ylabel("Total")
plt.show()
```

```python
# Graficamos la cantidad de titulos por clasificacion
plt.figure(figsize=(8,4))
sns.countplot(df, x="RATING")
plt.title("Cantidad de títulos por clasificación")
plt.xlabel("Clasificación")
plt.ylabel("Total")
plt.show()
```

```python
# Graficamos la clasificacion de los primeros 10 registros
plt.figure(figsize=(8,4))
sns.countplot(df.head(10), x="RATING")
plt.title("Clasificación de títulos (Muestra de 10 registros)")
plt.xlabel("Clasificación")
plt.ylabel("Total")
plt.show()
```

```python
# Graficamos los generos de los primeros 10 registros
plt.figure(figsize=(16,8))
sns.countplot(df.head(3), x="LISTED_IN")
plt.title("Géneros de títulos (Muestra de 3 registros)")
plt.xlabel("Categoría / Género")
plt.ylabel("Total")
plt.show()
```
