```python
import pandas as pd
```

```python
#Conjunto de datos
datos = {
    "influencer" : ["Ana","Luis","Martha", "Carolina"],
    "likes" : [],
    "comentarios" : [],
    "fecha" : [],
    "tipo_post" : []
}
```

```python
import random as rd
```

```python
cantidad = 1000000
```

```python
nombres = ["Ana","Luis","Martha", "Carolina","Sofia","Pedro","Elena","Diego"]
tipo = ["Foto","Video","Collage","Boomerang"]
```

```python
dfInsta = pd.DataFrame ({
    "influencer" : [rd.choice(nombres) for _ in range(cantidad)],
    "likes" : [rd.randint(0,1000) for _ in range(cantidad)],
    "comentarios" : [rd.randint(0,500) for _ in range(cantidad)],
    "fecha" : [rd.randint(2020,2026) for _ in range(cantidad)],
    "tipo_post" : [rd.choice(tipo) for _ in range(cantidad)]
})
```
