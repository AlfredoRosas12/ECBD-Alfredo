```python
import pandas as pd
import numpy as np
import random
```

```python
# Creación del conjunto de datos
datos = {
    "influencer": ["@DevCode", "@LinuxSetup", "@GamerTech", "@FlutterDev", "@ArchUser", "@Pythonista", "@WebDesign", "@TerminalTips"],
    "likes": [850, 4200, 3100, 1500, 2900, 5100, 1800, 3600],
    "comentarios": [24, 185, 95, 40, 110, 320, 55, 140],
    "fecha": ["2026-03-01", "2026-03-02", "2024-11-15", "2026-03-05", "2026-03-08", "2026-03-10", "2026-03-12", "2026-03-15"],
    "tipo_publicacion": ["carrusel", "reel", "foto", "historia", "reel", "carrusel", "foto", "historia"]
}
```

```python
# Creación del DataFrame inicial
df_tech = pd.DataFrame(datos)
```

```python
# Escalado del conjunto de datos a un millón de filas
dfMillon = pd.DataFrame(np.repeat(df_tech.values, 125000, axis=0), columns=df_tech.columns)
```

```python
# Conversión de tipos de datos para las columnas numéricas
dfMillon['likes'] = dfMillon['likes'].astype(int)
dfMillon['comentarios'] = dfMillon['comentarios'].astype(int)
```
